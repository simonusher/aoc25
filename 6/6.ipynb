```python
from operator import add, mul
from functools import reduce

opcode_to_op = {
    "*": mul,
    "+": add
}

def calc(opcode: str, args: list[int]) -> int:
    op = opcode_to_op[opcode]
    return reduce(op, args)

s = 0
for problem in zip(*arguments, operators):
    arguments = problem[:-1]
    opcode = problem[-1]
    s += calc(opcode, arguments)
s
```

```python
with open("6/input.txt") as f:
    inputt = f.read().strip("\n")
```

```python
lines = inputt.splitlines()
lines
```

```python
problems = []
buff = []
for x in zip(*lines):
    # print(x)
    if all(c == ' ' for c in x):
        problems.append(buff)
        buff = []
    else:
        buff.append(x)
problems.append(buff)
problems
```

```python
res = 0
for problem in problems:
    opcode = "".join(col[-1] for col in problem).strip()
    args = [int("".join(col[:-1]).strip()) for col in problem]
    res += calc(opcode, args)
res
```

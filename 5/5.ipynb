```python
def load(path: str):
    with open(path) as f:
        ranges_str, ingredients_str = f.read().split("\n\n")
        ranges = ranges_str.strip().splitlines()
        ingredients = [int(x) for x in ingredients_str.strip().splitlines()]
        ranges = [tuple(map(int, x.split("-"))) for x in ranges]
        return ranges, ingredients
```

```python
def is_fresh(ingredient: int, ranges: list[tuple[int, int]]) -> bool:
    return any(start <= ingredient <= end for start, end in ranges)
```

```python
def merge_ranges(ranges: list[tuple[int, int]], to_merge: list[tuple[int, int]]) -> list[tuple[int, int]]:
    if to_merge == []:
        return ranges
    
    *prev_ranges, prev_range = ranges
    next_range, *next_ranges = to_merge
    prev_range_start, prev_range_end = prev_range
    
    next_range_start, next_range_end = next_range

    if prev_range_start <= next_range_start <= prev_range_end:
        if next_range_end <= prev_range_end:
            to_return = [prev_range] # 2nd completely within first
        else:
            to_return = [(prev_range_start, next_range_end)]
    else:
        to_return = [prev_range, next_range]
    
    return merge_ranges(prev_ranges + to_return, next_ranges)


def fix_overlaps(ranges: list[tuple[int, int]]) -> list[tuple[int, int]]:
    ranges = sorted(ranges, key=lambda x: (x[0], -x[1]))

    first_range, *to_merge = ranges
    return merge_ranges([first_range], to_merge)
```

```python
def count_fresh(ranges: list[tuple[int, int]]):
    return sum(
        end - start + 1 for (start, end) in ranges
    )
```

```python
fixed_example = fix_overlaps(r_example)
fixed_example
```

```python
count_fresh(fixed_example)
```

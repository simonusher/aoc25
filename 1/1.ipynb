```python
DEFAULT_MAX_POSITION = 99

def turn(current_position: int, direction: str, distance: int, max_position: int | None = None) -> int:
    if max_position is None:
        max_position = DEFAULT_MAX_POSITION
    
    if direction == "L":
        distance = -distance
    
    return (current_position + distance) % (max_position + 1)
```

```python
turn(11, "R", 8)
```

```python
turn(19, "L", 19)
```

```python
def count(starting_position: int, instructions: list[str], watched_number: int) -> int:
    watched_count = 0

    current_position = starting_position

    for i in instructions:
        direction = i[0]
        distance = int(i[1:])
        current_position = turn(current_position, direction, distance)
        if current_position == watched_number:
            watched_count += 1
    return watched_count
```

```python
count(50, instructions, 0)
```

```python
MAX_POSITION = 99

def count2(starting_position: int, instructions: list[str], log=False) -> int:
    watched_count = 0

    current_position = starting_position

    for i in instructions:
        direction = i[0]
        distance = int(i[1:])
        num_rotations = distance // (MAX_POSITION+1)

        watched_count += num_rotations

        actual_distance = distance % (MAX_POSITION+1)
        previous_position = current_position
        
        current_position = turn(previous_position, direction, actual_distance)

        if current_position == 0 or (previous_position != 0 and ((direction == "L" and current_position > previous_position) or (direction == "R" and current_position < previous_position))):
            watched_count += 1
        
        if log:
            print(previous_position, direction, distance, actual_distance, current_position, watched_count)
    return watched_count
```

```python
example = [
    "L68",
    "L30",
    "R48",
    "L5",
    "R60",
    "L55",
    "L1",
    "L99",
    "R14",
    "L82",
]
```

```python
count2(50, example, log=True)
```

```python
count2(50, instructions, 0)
```

```python
def is_free(map: list[list[str]], y: int, x: int):
    return y < 0 or y >= len(map) or x < 0 or x >= len(map[0]) or map[y][x] == "."

def get_adjacent(y: int, x: int) -> list[int]:
    return [
        (y-1, x-1),
        (y-1, x),
        (y-1, x+1),
        (y, x-1),
        (y, x+1),
        (y+1, x-1),
        (y+1, x),
        (y+1, x+1),
    ]
```

```python
def load(path: str) -> list[str]:
    with open(path) as f:
        return [[c for c in l] for l in f.read().splitlines()]
```

```python
example = load("4/example.txt")
example
```

```python
def count_adjacent(map: list[list[str]], y: int, x: int) -> int:
    return sum(not is_free(map, y_a, x_a) for (y_a, x_a) in get_adjacent(y, x))
```

```python
count_adjacent(example, 0, 0)
```

```python
def get_accessible_rolls(map: list[list[str]]) -> list[tuple[int, int]]:
    access = []
    for y in range(len(map)):
        for x in range(len(map[0])):
            pos = map[y][x]
            if pos == "@":
                adjacent = count_adjacent(map, y, x)
                if adjacent < 4:
                    access.append((y, x))
    return access

def count_accesible_rolls(map: list[str]):
    return len(get_accessible_rolls(map))
```

```python
def remove_accessible(map: list[list[str]], accessible: list[tuple[int, int]]) -> list[list[str]]:
    for y, x in accessible:
        map[y][x] = "."
    return map
```

```python
def count_all_accesible(map: list[list[str]]):
    s = 0
    can_remove_anymore = True
    while can_remove_anymore:
        accessible = get_accessible_rolls(map)
        if len(accessible) == 0:
            can_remove_anymore = False
        else:
            s += len(accessible)
            map = remove_accessible(map, accessible)
    
    return s
```

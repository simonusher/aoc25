```python
def read_input(path: str):
    with open(path) as f:
        lines = f.read().splitlines()
        return lines
```

```python
def find_max_joltage(bank: str) -> int:
    index, first_digit = max(enumerate(bank[:-1]), key=lambda x: int(x[1]))
    second_digit = max(bank[index+1:], key=lambda x: int(x))
    return int(first_digit + second_digit)
```

```python
def total_joltage(banks: list[list[int]]):
    return sum(find_max_joltage(bank) for bank in banks)
```

```python
import functools

@functools.cache
def find_max_joltage(bank: str, need_to_leave=1) -> int:
    if need_to_leave > 0:
        to_search = bank[:-need_to_leave]
    else:
        to_search = bank
    index, digit = max(enumerate(to_search), key=lambda x: int(x[1]))
    return index, digit
```

```python
def find_max_joltage_multiple(bank: str, total_batteries=12) -> int:
    need_to_leave = total_batteries-1
    digits = []
    to_search = bank
    while need_to_leave >= 0:
        index, digit = find_max_joltage(to_search, need_to_leave)
        digits.append(digit)
        to_search = to_search[index+1:]
        need_to_leave-=1

    return int("".join(digits))
```

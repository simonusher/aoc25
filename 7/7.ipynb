```python
with open("7/example.txt") as f:
    example = [list(l) for l in f.read().splitlines()]
example
```

```python
def find_start(manifold: list[list[str]]) -> tuple[int, int]:
    for y, row in enumerate(manifold):
        for x, char in enumerate(row):
            if char == "S":
                return y, x

start_coords = find_start(example)
start_coords
```

```python
def count_splitters(manifold: list[list[str]], coords: tuple[int, int], used: set) -> int:
    if coords in used:
        return 0
    y, x = coords
    used.add(coords)
    if y >= len(manifold) or x >= len(manifold[0]) or x < 0:
        return 0
    
    char = manifold[y][x]
    if char == "S" or char == ".":
        return count_splitters(manifold, (y+1, x), used)
    
    elif char == "^":
        return 1 + count_splitters(manifold, (y, x-1), used) + count_splitters(manifold, (y, x+1), used)

    else:
        raise ValueError(f"Something wrong with coords: {coords}, char: {char}")
```

```python
def count(manifold: list[list[str]]) -> int:
    used = set()
    start_coords = find_start(manifold)
    return count_splitters(manifold, start_coords, used)
```

```python
count(example)
```

```python
import functools
def count_timelines(manifold: list[list[str]]) -> int:
    y_start, x_start = find_start(manifold)

    @functools.cache
    def _count_timelines(y: int, x: int) -> int:
        if y >= len(manifold) or x >= len(manifold[0]) or x < 0:
            return 1
        
        else:
            char = manifold[y][x]
            if char == "S" or char == ".":
                return _count_timelines(y+1, x)
            
            elif char == "^":
                return _count_timelines(y, x-1) + _count_timelines(y, x+1)

            else:
                raise ValueError(f"Something wrong with coords: {y}, {x}, char: {char}")

    return _count_timelines(y_start, x_start)
```

```python
count_timelines(example)
```

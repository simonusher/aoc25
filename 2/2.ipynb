```python
def is_invalid(id: int) -> bool:
    id_str = str(id)
    length = len(id_str)
    return id_str[:length//2] == id_str[length//2:]
```

```python
def load_ranges(path: str) -> list[tuple[int, int]]:
    with open(path, "r") as f:
        range_strs = f.read().split(",")
        ranges = [tuple(map(int, range_str.split("-"))) for range_str in range_strs]
        ranges
    return ranges
```

```python
ranges = load_ranges("2/example.txt")
ranges
```

```python
def sum_invalid(id_ranges: list[tuple[int,int]], invalid_func=is_invalid) -> int:
    s = 0

    for r_start, r_end in id_ranges:
        for i in range(r_start, r_end+1):
            if invalid_func(i):
                s += i
    return s
```

```python
import textwrap

def is_invalid2(id: int) -> bool:
    id_str = str(id)
    length = len(id_str)
    for i in range(1, length//2+1):
        if length % i == 0:
            if len(set(textwrap.wrap(id_str, i))) == 1:
                return True
    return False
```

```python
import re

pattern = re.compile(r'^(.+?)(?=\1+$)')

def is_invalid3(id: int) -> bool:
    id_str = str(id)

    return pattern.match(id_str) is not None
```

```python
ranges = load_ranges("2/input.txt")
sum_invalid(ranges, is_invalid3)
```

```python
def is_invalid4(id: int) -> bool:
    id_str = str(id)
    
    length = len(id_str)

    for n in range(1, length//2+1):
        if length % n == 0:
            chunks = [id_str[n*i:n*(i+1)] for i in range(length//n)]
            if len(set(chunks)) == 1:
                return True
    return False
    
```

```python
def is_invalid5(id: int) -> bool:
    s = str(id)
    return s in (s + s)[1:-1]
```

```python
ranges = load_ranges("2/input.txt")
sum_invalid(ranges, is_invalid5)
```
